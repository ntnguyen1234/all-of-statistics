```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [10, 5]
plt.rcParams["figure.autolayout"] = True

def f_eq(k):
    return np.exp(-(k+1))

def f_in(k):
    return 1/np.square(k)

ks = np.arange(2, 21)
plt.xticks(ks)

plt.plot(ks, f_eq(ks), color='red')
plt.plot(ks, f_in(ks), color='blue')
plt.legend(['Value', 'Chebyshev bound'])
plt.show()
```
