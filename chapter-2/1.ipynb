```python
from itertools import groupby

import matplotlib.pyplot as plt

data = list(range(7))
y_l = [0., 0., .1, .2, .2, 1., 1.]

x = []
y = []
# Get list (y) and index (x) of unique values in y_l
for k, g in groupby(range(len(y_l)), lambda idx: y_l[idx]):
    x.append(next(g))
    y.append(k)

fig, ax = plt.subplots()
ax.set_facecolor('white')

xmax = x[1:]
xmax.append(data[-1])

# https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.hlines.html
ax.hlines(y=y, xmin=x, xmax=xmax, color='red', zorder=1)

# https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.vlines.html
ax.vlines(x=x[1:], ymin=y[:-1], ymax=y[1:], color='red', linestyle='dashed', zorder=1)

ax.scatter(x[1:], y[1:], color='red', s=18, zorder=2)
ax.scatter(x[1:], y[:-1], color='white', s=18, zorder=2, edgecolor='red')
# ax.grid(False)
ax.set_xlim(data[0], data[-1])
ax.set_ylim([-0.02, 1.02])
```
