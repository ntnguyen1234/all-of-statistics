```python
import math
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [10, 5]
plt.rcParams["figure.autolayout"] = True

def f(x):
    return 1/np.sqrt(2*math.pi)*np.power(x, -(2+np.log(x))/2)

x = np.linspace(0, 5.5, 1000)
x[0] = 1e-6

plt.plot(x, f(x), color='red')
plt.show()
```

```python
import math
import numpy as np
from matplotlib import pyplot as plt

def f(u):
    return 1/np.sqrt(2*math.pi)*np.power(u, -(2+np.log(u))/2)

plt.rcParams["figure.figsize"] = [10, 5]
plt.rcParams["figure.autolayout"] = True
plt.xlim([-0.1, 10])

mu, sigma = 0, 1 # mean and standard deviation
x = np.random.normal(mu, sigma, 10000)
y = np.exp(x)

count, bins, ignored = plt.hist(y, 1000, density=True)
plt.plot(bins, f(bins), linewidth=2, color='r')
plt.show()
```
