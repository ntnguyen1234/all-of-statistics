```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [10, 5]
plt.rcParams["figure.autolayout"] = True
plt.xlim([-0.1, 5])

def f_1(u, beta: float):
    return -beta*np.log(1-u)

def f_exp(x, beta: float):
    return 1/beta*np.exp(-1/beta*x)

beta = 1
u = np.random.uniform(0,1,10000)
x = f_1(u, beta)
count, bins, ignored = plt.hist(x, 100, density=True)
plt.plot(bins, f_exp(bins, beta), linewidth=2, color='r')
plt.show()
```
