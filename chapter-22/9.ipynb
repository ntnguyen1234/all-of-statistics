```python
import numpy as np
from sklearn import datasets

iris: dict[str, ] = datasets.load_iris()

X: np.ndarray = iris['data']
Y: np.ndarray = iris['target']
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler


def get_accuracy(k: int) -> np.ndarray:
    clf = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=k))
    kf = StratifiedKFold()
    return cross_val_score(clf, X, Y, cv=kf, error_score='raise')
```

```python
from tqdm.notebook import tqdm

for k in tqdm(range(1, 121)):
    scores = get_accuracy(k)
    print(f'{k=} - Accuracy: {scores.mean():.3f} (+/- {scores.std():.3f})')
```
