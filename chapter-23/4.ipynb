```python
import numpy as np

rng = np.random.default_rng()
```

```python
a = .1
b = .3
N = 1_000_0

states = [1, 2]
P = np.array([[1-a, a],
              [b, 1-b]])
X = np.empty(N, dtype=np.int8)
X[0] = rng.choice(states)
```

```python
for i in range(1, N):
    X[i] = rng.choice(states, p=P[X[i-1]-1])
# print(X)

T = np.arange(1, N + 1)
pn_1 = np.cumsum(X == 1) / T
pn_2 = np.cumsum(X == 2) / T
```

```python
print(pn_1[-1], pn_2[-1])
```

```python
import matplotlib.pyplot as plt

plt.plot(T, pn_1)
plt.plot(T, pn_2)
plt.show()
```
