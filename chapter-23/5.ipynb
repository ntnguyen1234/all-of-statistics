```python
from functools import lru_cache


@lru_cache(maxsize=2)
def F(n):
    if n <= 1:
        return 0.
    else:
        return 1/4 * (F(n-1) + 1)**2
```

```python
import matplotlib.pyplot as plt

N = 1_000
x = list(range(1,N+1))

plt.figure(figsize=(12, 8))
plt.plot(x, list(map(F, x)))
plt.show()
```
