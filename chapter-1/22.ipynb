```python
from random import choices
import matplotlib.pyplot as plt

def mean(l: list[int | float]) -> float:
    """Return mean value of a list"""
    return sum(l)/len(l)

def coinflip(p: float, n: int, m: int) -> list[float]:
    """Return proportions of coin side with probability p tossing n times

    Args:
        p: Probability of getting a coin's side
        n: Number of tossing times
        m: Number of repeating experiments

    Returns:
        List of number of a coin's side over time
    """
    # Sample space
    space = [0, 1]
    probabilities = [1-p, p]

    # Store total number of heads after each experiment
    sum_x = []

    # Average values of total number of heads
    y = []

    for _ in range(m):
        x = choices(space, probabilities, k=n)
        sum_x.append(sum(x))
        y.append(mean(sum_x))

    return y

m = 500
p = 0.3
n = 10

y = coinflip(p, n, m)
plt.plot(list(range(1, m+1)), y)
plt.grid()
plt.show()
```
