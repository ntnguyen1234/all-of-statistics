```python
from random import choices
import matplotlib.pyplot as plt

def mean(l: list[int | float]) -> float:
    """Return mean value of a list"""
    return sum(l)/len(l)

def coinflip(p: float, n: int) -> list[float]:
    """Return proportions of coin side with probability p tossing n times

    Args:
        p: Probability of getting a coin's side
        n: Number of tossing times

    Returns:
        List of proportions of getting a coin's side over time
    """
    # Sample space, for example 1 is head, 0 is tail
    space = [0, 1]
    probabilities = [1-p, p]

    # List of {0, 1} values randomly generated by the probabilities above
    x = choices(space, probabilities, k=n)
    y = [mean(x[:i+1]) for i in range(n)]
    return y

n = 1000
p = 0.5

y = coinflip(p, n)
plt.plot(list(range(1, n+1)), y)
# plt.ylim([0, 1]) # Use this if you need to fix y-label limit from 0 to 1
plt.grid()
plt.show()
```
