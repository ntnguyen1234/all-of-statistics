```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [20, 5]
plt.rcParams["figure.autolayout"] = True

mu, sigma = 0, 1
x_bar = []

for n in range(1, 10001):
    x = np.random.normal(mu, sigma, n)
    x_bar.append(x.mean())

plt.scatter([*range(1, 10001)], x_bar, 2)
plt.show()
```
