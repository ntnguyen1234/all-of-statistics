```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [16, 16]
plt.rcParams["figure.autolayout"] = True
plt.xlim([-100, 10100])
plt.ylim([-0.2, 0.2])

mu, sigma = 0, 1
x_bar = []

for n in range(1, 10001):
    x = np.random.normal(mu, sigma, n)
    x_bar.append(x.mean())

plt.scatter(range(1, 10001), x_bar, 2)
plt.show()
```

```python
import numpy as np
from scipy.stats import norm

# Number of IID rv
n = 10000
ns = np.arange(1, n+1)

# Sample size
samples = 1

x_i = norm.rvs(size=(samples, n), random_state=42)
x_n = np.cumsum(x_i, axis=1) / ns
```

```python
from matplotlib import pyplot as plt

# Plot
plt.figure(figsize=(20, 10))

font_size = 20
label_pad = 30

for i in range(samples):
    plt.plot(ns, x_n[i])

plt.title('Normal distribution', fontsize=1.5*font_size, pad=label_pad)
plt.xlabel('Number of IID random variables', fontsize=font_size, labelpad=label_pad)
plt.ylabel('X_n', fontsize=font_size)

plt.xticks(fontsize=0.75*font_size)
plt.yticks(fontsize=0.75*font_size)

plt.xlim(-100, n+100)
plt.ylim(-0.5, 0.5)

plt.show()
```
