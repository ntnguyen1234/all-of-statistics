```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [10,5]
plt.rcParams["figure.autolayout"] = True
plt.xlim([-100, 10100])

x_n = [np.random.choice(np.array([-1,1], dtype=np.int8), size=n, replace=True).sum() for n in range(1, 10001)]
plt.scatter(range(1, 10001), x_n, 2)
plt.show()
```
