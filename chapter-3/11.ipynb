```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import bernoulli

# n IID rvs
n = 10000

# Number of samples
samples = 20

x_i = 2 * bernoulli.rvs(0.5, size=(samples, n), random_state=42) - 1
x_n = np.cumsum(x_i, axis=1)

ns = np.arange(1, n+1)

# Plot
plt.figure(figsize=(12, 8))

ns = np.arange(1, n+1)

for i in range(samples):
    plt.plot(ns, x_n[i])

plt.show()
```

```python
# Plot
plt.figure(figsize=(20, 10))

font_size = 20
label_pad = 30

for i in range(samples):
    plt.plot(ns, x_n[i])

plt.xlabel('Number of IID random variables', fontsize=font_size, labelpad=label_pad)
plt.ylabel('X_n', fontsize=font_size)

plt.xticks(fontsize=0.75*font_size)
plt.yticks(fontsize=0.75*font_size)

plt.xlim(-100, n+100)

plt.show()
```
